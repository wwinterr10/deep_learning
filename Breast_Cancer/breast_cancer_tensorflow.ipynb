```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
```

```python
#위스콘신 유방암 데이터를 읽어서 data에 저장
data = datasets.load_breast_cancer()
```

```python
#유방암 판단의 독립변수를 읽어서 X에 저장
X = data.data

#유방암 데이터의 종속 변수(악성: 0, 양성: 1)
y = data.target
```

```python
#X, y를 8:2로 나누어 훈련 데이터와 테스트 데이터에 대입
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler

#X에서 평균을 빼고 표준편차로 나누어 평균: 0, 표준편차: 1로 변환할 객체
scaler = StandardScaler()

#X_train의 평균과 표준편차를 계산
scaler.fit(X_train)
```

```python
#X_train에서 평균을 빼고 표준편차로 나누어 X_train에 대입
X_train = scaler.transform(X_train)
X_train
```

```python
#입력값을 읽어 예측할 Sequential 객체 생성
model = Sequential()

#Dense(출력데이터, input_dim=입력데이터, activation="활성함수")
#Dense 모델을 예측할 수 있도록 model에 추가
model.add(Dense(1, input_dim=30, activation="sigmoid"))
```

```python
#생성된 모델 정보 출력
model.summary()
```

```python
#Gradient Decent를 이용하여 w와 b를 찾는 방법 설정
model.compile(loss='binary_crossentropy', optimizer=Adam(lr=0.001), metrics=['acc'])
```

```python
#입력값을 X, 출력값을 y로 하여 100번 반복해서 w, b 계산
model.fit(X_train, y_train, epochs=100)
```

```python
#scaler를 이용해서 X_test의 범위를 변경
#X_test에서 X_train의 평균을 빼고 X_train의 표준편차로 나눠줌
X_test = scaler.transform(X_test)
```

```python
pred = model.predict(X_test)
```

```python
predict01 = np.where(pred > 0.5, 1, 0)
print("=" *100)
print("predict01")
print(predict01)
```

```python
#predict01을 1차원 배열로 변환
predict02 = predict01.flatten()
predict02
```

```python
#predict02와 y_test가 같으면 true, 다르면 false 리턴
predict03 = (predict02 == y_test)
predict03
```

```python
#true를 1로, false를 0으로 변환하여 합 계산
np.sum(predict03)
```

```python
#len(predict03): redict02의 줄의 수 리턴
len(predict03)
```

```python
#정확도 계산
acc = np.sum(predict03) / len(predict03)
acc
```
