```python
import numpy as np
```

```python
#AND 연산의 x1, x2 입력을 저장하는 배열
X = np.array([
        [0., 0.],
        [0., 1.],
        [1., 0.],
        [1., 1.]
    ], dtype = "float32")
```

```python
X
```

```python
#AND 연산의 y출력을 저장하는 배열
y = np.array([
        [0.],
        [0.],
        [0.],
        [1.]
    ], dtype = "float32")
```

```python
y
```

```python
#X에 곱해지는 값 선언
w = np.array([[1], [2]], dtype = "float32")

#X에 더해지는 값 선언
b = np.array([3], dtype = "float32")
```

```python
w
```

```python
b
```

```python
#X에 w를 곱하기
np.dot(X, w)
```

```python
#X에 w를 곱하고 b를 더하기
np.dot(X, w) + b
```

```python
#h를 0 ~ 1 사이 범위로 변경하는 활성함수 선언
def sigmoid(h):
    return 1 / (1 + np.exp(-h))
```

```python
#X * w + b를 실행해서 예측값을 hypothesis에 대입
hypothesis = np.dot(X, w) + b
hypothesis
```

```python
#hypothesis를 활성함수를 이용하여 값의 범위를 0 ~ 1 사이로 변경
hypothesis = sigmoid(np.dot(X, w) + b)
hypothesis
```

```python
#예측값 - 실제값
(hypothesis - y)
```

```python
#X의 줄칸을 반대로 전환
X.T
```

```python
#반대로 전환한 X의 0번째 줄을 리턴
X.T[0]
```

```python
#이차원 배열인 (hypothesis - y)와 변환하기 위해
#X.T[0]을 2차원 배열로 변환
X_col_2d1 = X.T[0].reshape(1, -1)
X_col_2d1
```

```python
#w1의 수정값
w1_gred = 1 / 4 * sum(np.dot(X_col_2d1, (hypothesis - y)))
print("w_gred:", w1_gred)
```

```python
#반대로 전환한 X의 1번째 줄을 리턴
X.T[1]
```

```python
#이차원 배열인 (hypothesis - y)와 변환하기 위해
#X.T[1]을 2차원 배열로 변환
X_col_2d2 = X.T[1].reshape(1, -1)
X_col_2d2
```

```python
#w2의 수정값
w2_gred = 1 / 4 * np.sum(np.dot(X_col_2d2, (hypothesis - y)))
print("w_gred:", w2_gred)
```

```python
#b의 수정값
b_gred = 1 / 4 * np.sum(hypothesis - y)
print("b_gred:", b_gred)
```

```python
#learning rate
learning_rate = 0.1

#1000번 반복해서 오차 제곱의 합이 0으로 수렴되는 w와 b를 찾음
for i in range(1000):
    for j in range(2):
        #예측값을 hypothesis에 대입
        hypothesis = sigmoid(np.dot(X, w) + b)
        print("hypothesis:", hypothesis)
        #오차를 계산
        cost = -1 / 4 * (y * np.log(hypothesis) + (1 - y) * np.log(1 - hypothesis))
        print("cost:", cost)
        #X의 줄칸을 변경하여 j번째 칸을 2차원 배열로 생성
        X_col_2d = X.T[j].reshape(1, -1)
        #w의 기울기 계산
        w_gred = 1 / 4 * np.sum(np.dot(X_col_2d, (hypothesis - y)))
        print("w_{}gred:{}".format(i, w_gred))
        #w 업데이트
        w[j] = w[j] - learning_rate * w_gred
        print("w{}:{}".format(j, w))
        #b의 기울기 계산
        b_gred = 1 / 4 * np.sum(hypothesis - y)
        print("b{}_gred:{}".format(i, b_gred))
        #b 업데이트
        b = b - learning_rate * b_gred
        print("b:{}".format(b))
```

```python
#Gradient Decent로 계산한 w
w
```

```python
#Gradient Decent로 계산한 b
b
```

```python
#X에 w를 곱하고 b를 더함
np.dot(X, w) + b
```

```python
#활성함수를 이용하여 데이터의 범위를 0 ~ 1로 설정
predict = sigmoid(np.dot(X, w) + b)
predict
```

```python
#True를 정수로 변환(1)
int(True)

#False를 정수로 변환(0)
int(False)
```

```python
#np.where(조건, 조건만족시 리턴값, 거짓일때 리턴값)
#predict에 저장된 값이 0.5 초과일 시 1 리턴, 아니면 0 리턴
predict01 = np.where(predict >= 0.5, 1, 0)
print("=" * 100)
print("predict01")
print(predict01)
```

```python
#predict01 == y가 같으면 True, 다르면 False 리턴
predict02 = (predict01 == y)
predict02
```

```python
#True는 1로, False는 0으로 변환하여 합을 계산
np.sum(predict02)
```

```python
#정확도 계산
acc = np.sum(predict02) / 4
acc
```
