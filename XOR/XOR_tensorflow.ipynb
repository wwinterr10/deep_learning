```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import numpy as np
```

```python
#XOR 연산의 x1, x2 입력을 저장하는 배열
X = np.array([
        [0., 0.],
        [0., 1.],
        [1., 0.],
        [1., 1.]
], dtype="float32")

#XOR 연산의 y 출력을 저장하는 배열
y = np.array([0, 1, 1, 0], dtype="float32")
```

```python
#입력값을 읽어서 예측할 Sequential 객체 생성
model = Sequential()

#모델에 데이터 분석 객체 추가
#Dense(10): 출력 데이터는 10번의 선형회귀를 실행
#input_dim=2: 입력 데이터의 칸 수는 2칸
#activation="sigmoid": 활성함수는 sigmoid를 사용
model.add(Dense(10, input_dim=2, activation="sigmoid"))
```

```python
#Dense(1): 출력 데이터는 1번의 선형회귀를 실행
#activation="sigmoid": 활성함수는 sigmoid를 사용
model.add(Dense(1, activation="sigmoid"))
```

```python
model.summary()
```

```python
#Gradient Desent를 이용해서 w0, b0, w1, b1을 찾는 방법 설정
model.compile(loss="binary_crossentropy", optimizer=Adam(lr=0.01), metrics=['acc'])
```

```python
#입력값을 X, 출력값을 y로 하고 500번 반복하여 w, b 계산
model.fit(X, y, epochs=500)
```

```python
#X에 대해 예측한 값을 pred에 대입
#pred에 값이 0.5 초과면 1, 아니면 0 리턴
pred = model.predict(X)
predict01 = np.where(pred > 0.5, 1, 0)
print(predict01)
```
