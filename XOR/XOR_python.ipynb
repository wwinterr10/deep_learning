```python
import numpy as np
```

```python
#XOR 연산의 x1, x2 입력을 저장하는 배열
X = np.array([
	[0., 0.],
	[0., 1.],
	[1., 0.],
	[1., 1]
], dtype="float32")
```

```python
#XOR 연산의 y 출력을 저장하는 배열
y = np.array([
	[0.],
	[1.],
	[1.],
	[0.]
], dtype="float32")
```

```python
#X에 곱해지는 값 선언
w = np.array([[1], [2]], dtype="float32")
```

```python
#X에 더해지는 값 선언
b = np.array([3], dtype="float32")
```

```python
#X에 w를 곱하고 b를 더한 예측값을 hypothesis에 대입
hypothesis = np.dot(X, w) + b
hypothesis
```

```python
#h를 0 ~ 1 사이 범위로 변경하는 활성함수 선언
def sigmoid(h):
	return 1 / (1 + np.exp(-h))
```

```python
#활성함수를 이용하여 hypothesis의 범위를 0 ~ 1 사이로 변경
hypothesis = sigmoid(np.dot(X, w) + b)
hypothesis
```

```python
#예측값 - 실제값
(hypothesis - y)
```

```python
#X의 줄칸을 반대로 전환하고 0번째 줄을 리턴
X.T
X.T[0]
```

```python
#2차원 배열인 (hypothesis - y)와 변환하기 위해 X.T[0]를 2차원 배열로 변환
X_col_2d1 = X.T[0].reshape(1, -1)
X_col_2d1
```

```python
#w1의 수정값
w1_gred = 1 / 4 * np.sum(np.dot(X_col_2d1, (hypothesis - y)))
print("w_gred:", w1_gred)
```

```python
#X의 줄칸을 반대로 전환하고 1번째 줄을 리턴
X.T[1]
```

```python
#2차원 배열인 (hypothesis - y)와 변환하기 위해 X.T[1]를 2차원 배열로 변환
X_col_2d2 = X.T[1].reshape(1, -1)
```

```python
#w2의 수정값
w2_gred = 1 / 4 * np.sum(np.dot((hypothesis - y), X_col_2d2))
print("w_gred:", w2_gred)
```

```python
#b의 수정값
b_gred = 1 / 4 * np.sum(hypothesis - y)
print("b_gred:", b_gred)
```

```python
#1000번 반복해서 오차 제곱의 합이 0으로 수렴되는 w, b를 찾음
learning_rate = 0.1

for i in range(1000):
    for j in range(2):
        #예측값을 hypothesis에 대입
        hypothesis = sigmoid(np.dot(X, w) + b)
        print("hypothesis:", hypothesis)

        #오차 계산
        cost = -1 / 4 * (y * np.log(hypothesis) + ( 1 - y) * np.log(1 - hypothesis))
        print("cost:", cost)

        #X의 줄칸을 변경하여 j번째 칸을 2차원 배열로 생성
        X_col_2d = X.T[j].reshape(1, -1)

        #w, b의 기울기를 계산하고 업데이트
        w_gred = 1 / 4 * np.sum(np.dot(X_col_2d, (hypothesis - y)))
        print("w_{}gred: {}".format(i, w_gred))
        w[j] = w[j] - learning_rate * w_gred
        print("w{} : {}".format(j, w))

        b_gred = 1 / 4 * np.sum(hypothesis - y)
        print("b{}_gred: {}".format(i, b_gred))
        b = b - learning_rate * b_gred
        print("b : {}".format(b))
```

```python
#X에 w를 곱하고 b를 더한 후 활성함수를 이용하여 데이터의 범위를 0 ~ 1로 설정
predict = sigmoid(np.dot(X, w) + b)
predict
```

```python
#predict에 저장된 값이 0.5 초과면 1, 아니면 0 리턴
predict01 = np.where(predict > 0.5, 1, 0)
print("=" * 100)
print("predict01")
print(predict01)
```

```python
#predict01과 y가 같으면 true, 다르면 false 리턴
predict02 = (predict01 == y)
predict02
```

```python
#true는 1로, false는 0으로 변환하여 합을 계산한 뒤 정확도를 계산
acc = np.sum(predict02) / 4
acc
```
