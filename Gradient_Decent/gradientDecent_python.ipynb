```python
import numpy as np
```

```python
#X는 독립변수로 입력값
X = np.array([1, 2, 3], dtype="float32")

#y는 종속변수로 출력값
y = np.array([2, 2.5, 3.5], dtype="float32")
```

```python
X
```

```python
y
```

```python
#초기 선형 회귀식은 wX + b = 2X + 1
w = 2
b = 1
```

```python
#X * W + b를 실행해서 예측값을 hypothesis에 대입
hypothesis = X * w + b
hypothesis
```

```python
#편미분하여 w, b의 기울기를 구하고 최소값으로 수렴하기 위해 편미분 값에 -1을 곱함
#Gradient Decent 편미분 공식에 의해 w_gred 계산
#np.dot(): 배열끼리 곱
w_gred = 1 / 3 * np.dot((hypothesis - y), X)
w_gred
```

```python
#w 업데이트
w = w - w_gred
w
```

```python
#Gradient Decent 편미분 공식에 의해 b_gred 계산
#np.sum(): 배열의 합
b_gred = 1 / 3 * np.sum(hypothesis - y)
b_gred
```

```python
#b 업데이트
b = b - b_gred
b
```

```python
#Gradient Decent 알고리즘을 반복 실행하여 오차제곱의 합이 최소화되는 w와 b를 찾음
for i in range(100):
    #예측값을 hypothesis에 대입
    hypothesis = X * w + b
    print("hypothesis:", hypothesis)
    #예측값 - 진짜값의 제곱의 합 coast를 계산
    MSE = np.sum((hypothesis - y) ** 2)
    print("MSE:", MSE)
    #w의 수정값
    w_gred = 1 / 3 * np.dot((hypothesis - y), X)
    print("w_gred:", w_gred)
    #b의 수정값
    b_gred = 1 / 3 * np.sum(hypothesis - y)
    print("b_gred:", b_gred)
    #w, b 업데이트
    w = w - w_gred
    print("w:", w)
    b = b - b_gred
    print("b:", b)
```

```python
#비용함수가 0에 가까운 값을 갖는 w, b를 계산하지 못함
#기울기인 w_gred와 b_gred가 너무 큰 값으로 계산되어 cost가 0이 되는 지점을 찾지 못함
#overshooting 현상 발생 -> overshooting이 일어나지 않도록 w_gred와 b_gred의 값을 감소시킴
```

```python
#w_gred와 b_gred에 1보다 작은 소수를 곱해서 w_gred, b_gred의 값 감소시킴
#learning rate: w_gred와 b_gred에 곱해지는 1보다 작은 소수값

#w와 b의 초기값
w = 2
b = 1
#w_gred와 b_gred에 곱할 1보다 작은 소수값 learning rate를 01.로 설정
learning_rate = 0.1

#1000번 반복해서 오차 제곱의 합이 0으로 수렴하는 w, b를 찾음
for i in range(1000):
    #예측값을 hypothesis에 대입
    hypothesis = X * w + b
    print("hypothesis:", hypothesis)
    #예측값 - 진짜값의 제곱의 합 coast를 계산
    MSE = np.sum((hypothesis - y) ** 2)
    print("MSE:", MSE)
    #w, b의 수정값
    w_gred = 1 / 3 * np.dot((hypothesis-y), X)
    print("w_gred:", w_gred)
    b_gred = 1 / 3 * np.sum(hypothesis - y)
    print("b_gred:", b_gred)
    #w, b 업데이트
    w = w - (learning_rate * w_gred)
    b = b - (learning_rate * b_gred)
    print("w:", w)
    print("b:", b)
```

```python
import matplotlib.pyplot as plt
```

```python
#X와 y값을 원모양 마커로 그래프를 그림
plt.plot(X, y, marker='o')

#X와 예측값 사이 붉은선(r), 선모양은 실선(-)으로 그래프를 그림
plt.plot(X, hypothesis, mfc='r', ls="-")

#y 좌표의 범위: 0 ~ 4
plt.ylim(0, 4)
```
